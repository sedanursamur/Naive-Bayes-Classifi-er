```python
import pandas as pd
from IPython.display import display
import math
```

```python
dataset = pd.read_csv('Training_Penguins_data.csv')
dataset
```

```python
#Separation of the training dataset into two sub-datasets based on the (sex) label
male = dataset[dataset['sex'] == 'male']
female = dataset[dataset['sex'] == 'female']

#P(male)
p_male = male.shape[0] / dataset.shape[0]

#P(Female)
p_female = female.shape[0] / dataset.shape[0]
display('P(male):', p_male,'P(female)',p_female)
```

```python
#homework 2.1.2
def discreteFeature(Name):
    discreteFeature = dataset[Name].unique()
    z = []
    for n in discreteFeature:
        a = male[male[Name] == n].shape[0]/male.shape[0]
        b = female[female[Name] == n].shape[0]/female.shape[0]
        z.append([a,b])
        
    distinct = pd.DataFrame(data = z,index = discreteFeature,columns= ['Probability_Male','Probability_Female'])
    display(distinct)
    return distinct

species = discreteFeature('species')
island = discreteFeature('island')
```

```python
def continuousFeatureCalculation(continuousColumnList):
    
    z = []
    for column in continuousColumnList:
        a = male[column].mean()
        b = female[column].mean()
        c = male[column].std()
        d = female[column].std()
        z.append([a,b,c,d])
        
    df_continuous = pd.DataFrame(data = z,index = continuousColumnList,columns= ['Male_Class_Mean','Female_Class_Mean','Male_SD','Female_SD'])
    display(df_continuous)
    return df_continuous
    
continiousFeatureTable = continuousFeatureCalculation(['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_kg'])    
```

```python
def calculate_Gauss_distribution(column):
    
    Gauss_Distribution=[]
    i=0
    sbt=3
  
    for column in column:
        for i in dataset[column]:

            i=((i-male[column].mean())**2)/(2*(male[column].std()**2))
            e=math.exp(-i)
            a1=(1/(male[column].std()*math.sqrt(2*sbt))) * e
            
            i=((i-female[column].mean())**2)/(2*(female[column].std()**2))
            e=math.exp(-i)
            a2=(1/(female[column].std()*math.sqrt(2*sbt))) * e
            
            Gauss_Distribution.append([a1,a2])
    
    continuous = pd.DataFrame(data = Gauss_Distribution, columns= ['Male','Female'])
    display(continuous)
    return continuous
```

```python
print('"bill_length_mm" feature’s values are distributing with Gaussian Distribution')
cft_bill_length_mm = calculate_Gauss_distribution(['bill_length_mm']) 

print('"bill_depth_mm" feature’s values are distributing with Gaussian Distribution')
cft_bill_depth_mm = calculate_Gauss_distribution(['bill_depth_mm']) 

print('"flipper_length_mm" feature’s values are distributing with Gaussian Distribution')
cft_flipper_length_mm = calculate_Gauss_distribution(['flipper_length_mm']) 

print('"body_mass_kg" feature’s values are distributing with Gaussian Distribution')
cft_body_mass_kg = calculate_Gauss_distribution(['body_mass_kg']) 
```

```python
dataset_test = pd.read_csv('Testing_Penguins_Data.csv')
y_test = dataset_test['sex']
dataset_test
```

```python
#calculate male score

a1 = 0
a2 = 0
a3 = 0
a4 = 0
a5 = 0
a6 = 0
predict = []

for row in dataset_test.iterrows():
    for row1 in species.iterrows():
        if (row[1][0] == row1[0]):
            a1 = row1[1][0]
    for row1 in island.iterrows():
        if (row[1][1] == row1[0]):
             a2 = row1[1][0]
    
    for column in male.columns:
        for row1 in continiousFeatureTable.iterrows():
            if column == row1[0]:
                if column == 'bill_length_mm':
                    a3 = (1/(math.sqrt(2*3*(row1[1][2]**2))))*(math.e**((-(row[1][2]-row1[1][0])**2)/(2*(row1[1][2]**2))))
                if column == 'bill_depth_mm':
                    a4 = (1/(math.sqrt(2*3*(row1[1][2]**2))))*(math.e**((-(row[1][3]-row1[1][0])**2)/(2*(row1[1][2]**2))))
                if column == 'flipper_length_mm':
                    a5 = (1/(math.sqrt(2*3*(row1[1][2]**2))))*(math.e**((-(row[1][4]-row1[1][0])**2)/(2*(row1[1][2]**2))))
                if column == 'body_mass_kg':
                    a6 = (1/(math.sqrt(2*3*(row1[1][2]**2))))*(math.e**((-(row[1][5]-row1[1][0])**2)/(2*(row1[1][2]**2))))
    maleScore = p_male * a1 * a2 * a3 * a4 * a5 * a6
           
    for row1 in species.iterrows():
        if (row[1][0] == row1[0]):
            a1 = row1[1][1]
    for row1 in island.iterrows():
        if (row[1][1] == row1[0]):
             a2 = row1[1][1]
    
    for column in male.columns:
        for row1 in continiousFeatureTable.iterrows():
            if column == row1[0]:
                if column == 'bill_length_mm':
                    a3 = (1/(math.sqrt(2*3*(row1[1][3]**2))))*(math.e**((-(row[1][2]-row1[1][1])**2)/(2*(row1[1][3]**2))))
                if column == 'bill_depth_mm':
                    a4 = (1/(math.sqrt(2*3*(row1[1][3]**2))))*(math.e**((-(row[1][3]-row1[1][1])**2)/(2*(row1[1][3]**2))))
                if column == 'flipper_length_mm':
                    a5 = (1/(math.sqrt(2*3*(row1[1][3]**2))))*(math.e**((-(row[1][4]-row1[1][1])**2)/(2*(row1[1][3]**2))))
                if column == 'body_mass_kg':
                    a6 = (1/(math.sqrt(2*3*(row1[1][3]**2))))*(math.e**((-(row[1][5]-row1[1][1])**2)/(2*(row1[1][3]**2))))

    femaleScore = p_female * a1 * a2 * a3 * a4 * a5 * a6
    if maleScore < femaleScore:
        predict.append('female')
    else:
        predict.append('male')       
        
print(predict)
```

```python
#calculate total acc.
total_accuarcy = 0
for i in range (len(predict)):
    if predict[i] == y_test[i]:
        total_accuarcy = total_accuarcy + 1
        
total_accuarcy = total_accuarcy / len(predict)

print('Total Accuracy : ', total_accuarcy)

# calculate male acc.
test_male = dataset_test[dataset_test['sex'] == 'male']
test_female = dataset_test[dataset_test['sex'] == 'female']

male_accuarcy = 0
malePositive = 0
for i in range (len(predict)):
    if predict[i] == 'male':
        if predict[i] == y_test[i]:
            male_accuarcy = male_accuarcy + 1
            malePositive = malePositive + 1



male_accuarcy = male_accuarcy / len(test_male)


print('Male Accuracy : ', male_accuarcy)

#calculate female acc.
female_accuarcy = 0
femalePositive = 0
for i in range (len(predict)):
    if predict[i] == 'female':
        if predict[i] == y_test[i]:
            female_accuarcy = female_accuarcy + 1
            femalePositive = femalePositive + 1



female_accuarcy = female_accuarcy / len(test_female)


print('Female Accuracy : ', female_accuarcy)

        
```

```python
femaleNegative = test_female.shape[0] - femalePositive
maleNegative = test_male.shape[0] - malePositive
confData = []

confData.append([malePositive,femaleNegative])
confData.append([maleNegative,femalePositive])

confMatrix = pd.DataFrame(data = confData,index = ['Predicted Male', 'Predicted Female'],columns= ['Actual Male','Actual Female'])

confMatrix
```
